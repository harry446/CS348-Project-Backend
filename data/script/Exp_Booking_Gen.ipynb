## Booking

```python
spot_df = pd.read_csv("allspots.txt", header=None)
spot_df.columns = ['lid', 'sid', 'parking_type', 'latitude', 'longitude', 'max_stay', 'price']
lot_df = pd.read_csv("lots.txt", header=None)
lot_df.columns = ['lid', 'area', 'address', 'lot_name', 'capacity', 'like_num']
p_holders_df = pd.read_csv("p_holders.txt", header=None)
p_holders_df.columns = ['uid', 'type', 'expiry_date'] 
```

```python
permit_spot_df = spot_df[spot_df['parking_type'] == 'permit'].copy().reset_index(drop=True)
accessible_spot_df = spot_df[spot_df['parking_type'] == 'accessible'].copy().reset_index(drop=True)
pay_spot_df = spot_df[spot_df['parking_type'] == 'pay'].copy().reset_index(drop=True)
free_spot_df = spot_df[spot_df['parking_type'] == 'free'].copy().reset_index(drop=True)
```

```python
permit_user_df = user_df[user_df['uid'].isin(p_holders_df['uid'])].copy().reset_index(drop=True)
acc_user_df = user_df[user_df['is_accessible'] == 1].copy().reset_index(drop=True)
permit_user_df = permit_user_df[~permit_user_df['uid'].isin(acc_user_df['uid'])].copy().reset_index(drop=True)
vis_user_df = user_df[user_df['identity'] == "visitor"].copy().reset_index(drop=True)
```

```python
bookings = []
start_d = datetime(2023, 1, 1, 8, 0, 0)
start_d_str = start_d.strftime('%Y-%m-%d %H:%M:%S')
create_time = start_d_str

start_t = start_d + timedelta(hours=0.5)
start_time = start_t.strftime('%Y-%m-%d %H:%M:%S')
```

```python
# accessible
for i in range(len(accessible_spot_df)):
    duration = accessible_spot_df['max_stay'][i]
    end_t = start_t + timedelta(hours = duration)
    end_time = end_t.strftime('%Y-%m-%d %H:%M:%S')
    bookings.append(
        (i + 1, acc_user_df['uid'][i], accessible_spot_df['lid'][i], accessible_spot_df['sid'][i], 
         create_time, start_time, end_time, accessible_spot_df['price'][i], 
         random.choices(population = [0, 1], weights = [0.5, 0.5], k = 1)[0], 0)
    )
```

```python
# pay
for i in range(len(pay_spot_df)):
    duration = pay_spot_df['max_stay'][i]
    end_t = start_t + timedelta(hours = duration)
    end_time = end_t.strftime('%Y-%m-%d %H:%M:%S')
    bookings.append(
        (i + 1 + len(accessible_spot_df), acc_user_df['uid'][i], pay_spot_df['lid'][i], pay_spot_df['sid'][i], 
         create_time, start_time, end_time, pay_spot_df['price'][i], 
         random.choices(population = [0, 1], weights = [0.5, 0.5], k = 1)[0], 0)
    )
```

```python
# permit
for i in range(len(permit_spot_df)):
    duration = 168
    end_t = start_t + timedelta(hours = duration)
    end_time = end_t.strftime('%Y-%m-%d %H:%M:%S')
    bookings.append(
        (i + 1 + len(accessible_spot_df) + len(pay_spot_df), permit_user_df['uid'][i], permit_spot_df['lid'][i], permit_spot_df['sid'][i], 
         create_time, start_time, end_time, permit_spot_df['price'][i], 
         random.choices(population = [0, 1], weights = [0.5, 0.5], k = 1)[0], 0)
    )
```

```python
# free
for i in range(len(free_spot_df)):
    duration = free_spot_df['max_stay'][i]
    end_t = start_t + timedelta(hours = duration)
    end_time = end_t.strftime('%Y-%m-%d %H:%M:%S')
    bookings.append(
        (i + 1 + len(accessible_spot_df) + len(pay_spot_df) + len(permit_spot_df), acc_user_df['uid'][i], free_spot_df['lid'][i], free_spot_df['sid'][i], 
         create_time, start_time, end_time, free_spot_df['price'][i], 
         random.choices(population = [0, 1], weights = [0.5, 0.5], k = 1)[0], 0)
    )
```

```python
booking_df = pd.DataFrame(bookings, 
                          columns = ['bid', 'uid', 'lid', 'sid', 'create_time', 'start_time', 'end_time', 'price', 'status', 'liked'])
```

```python
booking_df.to_csv('expired_booking.txt', index=False, header=False, lineterminator='\n', na_rep = 'NULL')
```
